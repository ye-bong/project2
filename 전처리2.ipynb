# 다변량 데이터 분석 시나리오

## 분석 주제:
- 해법수학이 진짜로 잘나가는 이유가 무엇인가 분석

## 분석 목표:
- 다변량 시각화를 한다.
- 향후 추세 예측을 한다.
- 경쟁사와 비교 분석을 한다.

1. 해법수학 입지 정보 :크롤링

```python
# 면적대비 공시가격 알아보기
price['공시가격비율'] = price['공시가격'] / price['전용면적']
```
